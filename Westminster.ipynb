## Best In Show, Westminster

```python
# Importing libraries needed for this project
import os
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
```

### Setting up CSV file

```python
#importing csv file to jupyter notebook
dog_show = pd.read_csv("BestInShow.csv")
```

```python
dog_show
```

### Creating an sqlite database

```python
dog_show.to_sql('dog_show', sqlite3.connect('dog_show.db'), if_exists = "replace")
```

### Connecting to sqlite

```python
con = sqlite3.connect('dog_show.db')
dog_show
```

### What breed of dog has won the most from the beginning of the competition?

```python
breed_winners = pd.read_sql_query("SELECT Breed, COUNT(Breed) AS counted_breeds FROM dog_show GROUP BY Breed", con);
```

```python
breed_winners
```

### Now that the different breeds are counted, it's time to graph them.

```python
my_plot = breed_winners.plot(kind='bar', x='Breed', y='counted_breeds', legend=None,title="Which breed has won the most",figsize=(12, 10))
my_plot.set_xlabel("Breed")
my_plot.set_ylabel("Total number of wins")
```

## Fox Terriors has won the most competitions. 

## How many owner's have won more than one competition?

```python
owner_winners = pd.read_sql_query("SELECT Winner, COUNT(Winner) AS counted_winners FROM dog_show GROUP BY Winner HAVING COUNT(*) >1", con);
```

```python
owner_winners
```

### Let's graph these results...

```python
my_plot = owner_winners.plot(kind='bar', x='Winner', y='counted_winners', legend=None,title="Which winner won the most",figsize=(10, 8))
my_plot.set_xlabel("Winner")
my_plot.set_ylabel("Total number of wins")
```

### 7 owners have more than 1 win.

### Which category has the most wins in it?

```python
category_winners = pd.read_sql_query("SELECT Category, COUNT(Category) AS Type_of_dog FROM dog_show GROUP BY Category", con);
```

```python
category_winners
```

```python
plt.figure(figsize = (10, 10))
x = category_winners.Type_of_dog
plt.pie(x)
plt.legend(labels = category_winners.Category, loc='upper right')
plt.show()
```

### Terrier's have the most wins.
